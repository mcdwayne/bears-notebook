```python
import requests
import random
import time
from IPython import display


bearwidth = random.randrange(100, 1000)
bearheight = random.randrange(100, 1000)
response = requests.get("https://placebear.com/{}/{}.jpg".format(bearwidth, bearheight))
bear_url = response.url 
print(bear_url)
# time.sleep(5)
display.Image(bear_url)
    
```
